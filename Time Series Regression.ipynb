```python
import numpy as np
import pandas as pd
import datetime
timeseries = pd.read_csv('data.csv')
```

```python
timeseries.head(10)
```

```python
timeseries = timeseries.drop('Unnamed: 0',axis=1)
```

```python
timeseries.dtypes
```

```python
timeseries.columns
```

```python
timeseries.columns = timeseries.columns.str.replace(' ', '')
```

```python
timeseries.columns
```

```python
train = timeseries.iloc[0:1464,:]
```

```python
train
```

```python
test = timeseries.iloc[1464:,:]
```

```python
test
```

```python
test.drop(['datetime'],axis = 1 , inplace= True)
```

```python
test.columns.tolist()
```

```python
train.drop(['datetime'],axis = 1 , inplace= True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
```

# next 6 hour

##  all

```python
features = []
target = []
for i in range(len(train)-6):
    features.append(train.iloc[i:(i+6)].T)
    target.append(train.iloc[i+6])


features_t = []
target_t= []
for i in range(len(test)-6): 
    features_t.append(test[i:(i+6)].T)
    target_t.append(test.iloc[i+6])


#LinearRegression
linear = LinearRegression()  
linear.fit(features[i], target[i])
linear_pre = linear.predict(features_t[i])
print("MSE of linear regresssion to predict all features by next 12 hours",metrics.mean_absolute_error(target_t[i], linear_pre))
#ramdon forest Regression
forest_reg = RandomForestRegressor()
forest_reg.fit(features[i], target[i])
forest_reg_pred = forest_reg.predict(features_t[i])
print("MSE of random forest regresssion to predict all features by next 12 hours",metrics.mean_absolute_error(target_t[i], forest_reg_pred))
```

## PM2.5 

```python
PM_train= train['PM2.5']
PM_test = test['PM2.5']



features = []
target = []
for i in range(len(PM_train)-6):
    features.append(PM_train.iloc[i:(i+6)])
    target.append(PM_train.iloc[i+6])


features_t= []
target_t= []
for i in range(len(PM_test)-6): 
    features_t.append(PM_test[i:(i+6)])
    target_t.append(PM_test.iloc[i+6])


#linear model
linear = LinearRegression()  
linear.fit(features, target)
linear_pre = linear.predict(features_t)
print("MSE of linear regresssion to predict PM2.5 by next 6 hours ",metrics.mean_absolute_error(target_t, linear_pre))
# ramdon forest Regression
forest_reg = RandomForestRegressor()
forest_reg.fit(features, target)
forest_reg_pred = forest_reg.predict(features_t)
print("MSE of random forest regresssion to predict PM2.5 by next 6 hours",metrics.mean_absolute_error(target_t, forest_reg_pred))
```

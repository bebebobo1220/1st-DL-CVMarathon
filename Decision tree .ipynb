# 載入fuction 讀取data 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('character-deaths.csv')
```

# preprocessing 資料前處理

### 1.選擇 Death Year 為標籤，並將空值轉為0,有數值轉為1

```python
ma=df['Death Year']
ma_na=np.where(np.isnan(ma)==True,0,1)
df['Death Year']=pd.DataFrame(ma_na)
df
```

### 2.其他欄位空值補 0

```python
df =df.fillna(0)
```

### 3. 將Allegiances 欄位轉為標籤編碼 再進行獨熱編碼(dummy)

```python
labelencoder = LabelEncoder()
sub_df = pd.DataFrame(df['Allegiances'])
print(sub_df.shape)
sub_df.head()
df[['Allegiances']]=labelencoder.fit_transform(sub_df)
```

```python
df_dummy= pd.get_dummies(df,columns=['Allegiances'])
df_dummy
```

### 4.1 將 'Name'Book of death ''Death Chapter' 欄位刪除，並設為目標值

```python
train_X = df_dummy.drop(df_dummy.columns[[0,1,2,3,4]],axis=1) 
train_X
```

### 4.2選定特徵值

```python
train_Y = df[['Death Year']]
```

### 5.亂數拆成訓練集(75%)與測試集(25%) 

```python
x_train, x_test, y_train, y_test = train_test_split(train_X,train_Y ,test_size=0.25, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

# 使用scikit-learn的DecisionTreeClassifier進行預測

```python
# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
classifier = clf.fit(x_train, y_train) 

y_pred = clf.predict(x_test)
```

# 做出Confusion Matrix，並計算Precision, Recall, Accuracy

### accuracy

```python
from sklearn import datasets, metrics

acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

### confusion_matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

### recall

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred, average='macro')
```

```python
recall_score(y_test, y_pred, average='weighted')
```

### precision


```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred, average='macro')
```
